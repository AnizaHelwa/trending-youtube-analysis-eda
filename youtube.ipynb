#### Tentang Dataset
Dataset YouTube New (Trending YouTube Video Statistics) dari Kaggle ini sebenarnya adalah rekaman harian dari video-video yang masuk ke daftar Trending di berbagai negara.

Daftar Trending di YouTube itu unik karena tidak hanya ditentukan oleh jumlah views terbanyak, tapi juga kombinasi dari seberapa cepat views meningkat (momentum), asal views, dan interaksi penonton.

1. Cakupan Negara: Dataset ini biasanya mencakup data dari 10 wilayah/negara berbeda: Amerika Serikat (US), Kanada (CA), Jerman (DE), Perancis (FR), Inggris (GB), India (IN), Jepang (JP), Korea Selatan (KR), Meksiko (MX), dan Rusia (RU).
2. Informasi yang Tersedia (Kolom Utama): 
   - Identitas Video: Judul (title), ID video, nama channel, waktu rilis (publish_time), dan tag.
   - Statistik Performa: Jumlah penonton (views), jumlah suka (likes), jumlah tidak suka (dislikes), dan jumlah komentar (comment_count).
   - Kategori: Diwakili oleh category_id.
   - Waktu Trending: Tanggal saat video tersebut muncul di tab trending (trending_date).
   - Status Interaksi: Apakah komentar atau rating-nya dimatikan oleh pemilik channel (comments_disabled, ratings_disabled).
